#### 75. How to compute averages using a sliding window over an array? (★★★) 
(**hint**: np.cumsum)

````python
import numpy as np
def moving_average(a, n) :
    test = np.cumsum(a, dtype=float)
    test[n:] = test[n:] - test[:-n]
    return test[n - 1:] / n
moving_average(np.arange(20),5)
````
